```python
# Sigmoid baby
def sigge(x):
    return 1/(1+np.exp(-x))
```

```python
def initializeParameters(dim, output):
    """initilize the parameters w and b, returns w and b """
    w = np.random.randn(dim, output)* np.sqrt(2/dim) # * 0.001
    b = np.zeros((1, output))
    assert (w.shape == (dim, output))

    return w, b
```

```python
def MatrixMaker(matrix):
    """ Returns 4 values: X - Matrix, y - vector, parameter (weight), parameter (bias) """
    y = matrix['class'].values                   # Truth Label
    y = y.reshape(-1, 1)
    X = matrix.drop(['0' ,'class'], axis = 1)

    return X, y #w, b
```

```python
def updateParameters(w, input, output):
    w = np.random.randn(input, output) * 0.001
    return w
```

```python
def nn_linear(X, w, b):
    #w = np.random.randn(Input,Output)
    A = np.dot(X,w)+b 
    return A
```

```python
def reLU(z):
    return np.maximum(0, z)
```

```python
# DEBUGGER
X, y = MatrixMaker(training_set)
print(y.shape)
#grads, cost = propagate(w, b, X, y)
#print(cost)
```

```python
def forward(w1, w2, b1, b2, X):
    """ All the beautiful little layers produced """
    Z1 = nn_linear(X, w1, b1)
    A1 = reLU(Z1)
    Z2 = nn_linear(A1, w2, b2)
    A2 = sigge(Z2)

    return Z1, Z2, A1, A2, w2
```

```python
def backward(w2, A1, A2, Z1, X, Y):
    """ My little beautiful derivatives and backwards chain ruuule """
    m = X.shape[0]

    dZ2 = A2 - Y
    dw2 = (1/m) * np.dot(A1.T,dZ2)
    db2 = (1/m) * np.sum(dZ2, axis = 0, keepdims=True)
    
    dA1 = np.dot(dZ2, w2.T)
    dZ1 = (dA1 * (Z1 > 0)).astype(float)
    dw1 = (1/m) * np.dot(X.T, dZ1)
    db1 = (1/m) * np.sum(dZ1, axis = 0, keepdims=True)
    
    grads = {"dw1": dw1,
             "dw2": dw2,
             "db1": db1,
             "db2": db2}
    
    return grads
```

```python
# GRADED FUNCTION: predict

def predict(w1, w2, b1, b2, X):
   
    _, _, _, A, _ = forward(w1, w2, b1, b2, X)
    
    Y_prediction = np.where(A>0.5, 1, 0)

    return Y_prediction
```

```python
# GRADED FUNCTION: optimize

def optimize(X, Y, num_iterations, learning_rate, print_cost = False):


    m = X.shape[0]
    input_size = X.shape[1]
    hidden_size = 4
    output_size = 1
    w1, b1 = initializeParameters(input_size, hidden_size)  #18x4?
    w2, b2 = initializeParameters(hidden_size, output_size) #4x1
    
    train_accuracy_history = []
    test_accuracy_history = []
    value_accuracy_history = []
    costs = []
    Xtrain, ytrain = MatrixMaker(training_set)
    Xtest, ytest = MatrixMaker(test_set)
    Xval, yval = MatrixMaker(value_set)
    epsilon = 1e-8  # Small value to avoid log(0)

    for i in range(num_iterations):

        Z1, Z2, A1, A2, w2 = forward(w1, w2, b1, b2, X)

        # BCE - Binary Cross Entropy
        cost = (-1/m) * np.sum((Y * np.log(A2 + epsilon)) + ((1-Y) * (np.log(1-A2+epsilon))) ) 

        # Retrieve derivatives from grads

        grads = backward(w2, A1, A2, Z1, X, Y)

        dw1 = grads["dw1"]
        dw2 = grads["dw2"]
        db1 = grads["db1"]
        db2 = grads["db2"]

        # gradient decsent
        w1 = w1 - learning_rate * dw1
        #print("weight 1: ", w1)
        w2 = w2 - learning_rate * dw2
        #print("weight 2: ", w2)
        b1 = b1 - learning_rate * db1
        b2 = b2 - learning_rate * db2

        # Record the costs
        costs.append(cost)

        Y_prediction_train = predict(w1, w2, b1, b2, Xtrain)
        Y_prediction_validation = predict(w1, w2, b1, b2, Xval)
        train_accuracy = (100 - np.mean(np.abs(Y_prediction_train - ytrain)) * 100)
        validation_accuracy = (100 - np.mean(np.abs(Y_prediction_validation - yval)) * 100)
        train_accuracy_history.append(train_accuracy)
        value_accuracy_history.append(validation_accuracy)

        # Print the cost every 100 training examples
        if print_cost and i % 100 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
            

    params = {"w1": w1,
              "w2": w2,
              "b1": b1,
              "b2": b2}

    grads = {"dw1": dw1,
             "dw2": dw2,
             "db1": db1,
             "db2": db2}
    #test_accuracy_history,
    return params, grads, costs, train_accuracy_history, value_accuracy_history
```

```python
# GRADED FUNCTION: model

def model(training_set,test_set, num_iterations = 200, learning_rate = 0.5, print_cost = False):
    
    Xtrain, ytrain = MatrixMaker(training_set)
    Xtest, ytest = MatrixMaker(test_set)
    
     #test_history,
    parameters, _, costs, train_history, vali_history = optimize(Xtrain, ytrain, num_iterations, learning_rate, print_cost)

    # Retrieve parameters w and b from dictionary "parameters"
    w1 = parameters['w1']
    w2 = parameters['w2']
    b1 = parameters["b1"]
    b2 = parameters['b2']

    # Predict test/train set examples 
    Y_prediction_test = predict(w1, w2, b1, b2, Xtest)
    Y_prediction_train = predict(w1, w2, b1, b2, Xtrain)

    #print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - ytrain)) * 100))
    print("-"*30)
    print("train accuracy: {} %".format(train_history[-1]))
    print("validation accuracy: {} %".format(vali_history[-1]))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - ytest)) * 100))


    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test,
         "Y_prediction_train" : Y_prediction_train,
         "w" : w2,
         "b" : b2,
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
     
    # Plot the training and validation loss over epoch
    plt.figure(figsize=(12, 5))

    # Loss Plot
    plt.subplot(1, 2, 1)
    plt.plot(range(num_iterations), costs, label='Training Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss')
    plt.legend()

    # Accuracy Plot
    plt.subplot(1, 2, 2)
    plt.plot(range(num_iterations), np.array(train_history), label='Training Accuracy')
    #plt.plot(range(num_iterations), np.array(test_history), label='Test Accuracy')
    plt.plot(range(num_iterations), np.array(vali_history), label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.title('Training Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()
    
    return d
```

```python
d = model(training_set, test_set, num_iterations = 1000, learning_rate = 0.1, print_cost = True)
```
